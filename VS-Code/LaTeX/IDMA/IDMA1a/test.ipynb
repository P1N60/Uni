```python
# Re-import necessary libraries after execution state reset
import matplotlib.pyplot as plt
import numpy as np
import math

# Re-define the search function
def search(A, lo, hi):
    if A[lo] >= A[hi]:
        return "failure"
    elif lo + 1 == hi:
        return lo
    else:
        mid = math.floor((lo + hi) / 2)
        if A[mid] > A[lo]:
            return search(A, lo, mid)
        else:
            return search(A, mid, hi)

# Function to plot sequences
def plot_sequence_with_zero(A, title):
    indices = np.arange(0, len(A))  # Include index 0
    values = [0] + A[1:]  # Explicitly set index 0 to 0

    plt.figure(figsize=(6, 4))
    plt.plot(indices, values, marker='o', linestyle='-', color='b', label="Sequence")
    plt.xlabel("Index")
    plt.ylabel("Value")
    plt.title(title)
    plt.grid(True)

    # Highlight local minimum found by the algorithm
    min_index = search(A, 1, len(A) - 1)
    if isinstance(min_index, int):  # Ensure it’s not "failure"
        plt.scatter(min_index, A[min_index], color='red', zorder=3, label="Local Minimum")

    plt.legend()
    plt.show()

# Example increasing-decreasing sequences
example_arrays = [
    [0, 5, 100, 8, 7, 6, 1, 4, 10],   # User's example
    [0, 1, 2, 3, 4, 5, 6, 7, 8],       # Strictly increasing
    [0, 8, 7, 6, 5, 4, 3, 2, 1],       # Strictly decreasing
    [0, 3, 2, 1, 4, 5, 6, 7, 8],       # Valley at index 3
    [0, 10, 5, 2, 3, 6, 8, 9, 12],      # Valley at index 3 or 4
    [0, 20, 15, 10, 5, 1, 2, 3, 4],   # Clear valley at index 5 (1 is min)
    [0, 7, 6, 5, 4, 3, 2, 3, 4],      # Valley at index 6 (2 is min)
    [0, 10, 9, 8, 7, 6, 7, 8, 9],     # Valley at index 6 (6 is min)
    [0, 50, 40, 30, 20, 10, 15, 25],  # Valley at index 5 (10 is min)
    [0, 2, 3, 5, 8, 10, 9, 7, 4, 1],  # Valley at index 9 (1 is min)
]

# Plot each example
for i, A in enumerate(example_arrays):
    plot_sequence_with_zero(A, f"Example {i+1}")
```
